### Câu 1: Biến là gì?

**Câu hỏi:** Trong Blockly, biến (variable) là gì?

**Các lựa chọn:**
- A) Một loại vòng lặp
- B) Một hộp chứa có tên để lưu giá trị (số, chữ, danh sách, v.v.)
- C) Một khối vẽ hình
- D) Một toán tử so sánh
- E) Một hàm chỉ chạy một lần

**Đáp án: B**

---

### Giải thích chi tiết:

**1. Khái niệm biến:**
- Biến là **nơi lưu trữ dữ liệu** có tên (ví dụ: `điểm`, `tên`, `danh_sách`)
- Biến có thể chứa số, chữ (text), danh sách (list), hoặc giá trị đúng/sai (boolean)
- Giá trị trong biến có thể **thay đổi** trong quá trình chương trình chạy

**2. Ví dụ:**
- `set [điểm] to [10]` → biến "điểm" lưu số 10
- `set [tên] to ["Lan"]` → biến "tên" lưu chuỗi "Lan"

**3. Tại sao các đáp án khác sai:**
- A: Vòng lặp dùng để lặp lại hành động, không phải lưu giá trị
- C, D, E: Không mô tả đúng biến

**Ghi nhớ:** Biến = hộp có tên để chứa dữ liệu, có thể đọc và ghi (thay đổi) giá trị.

### Câu 2: Khối Lặp (Repeat)

**Câu hỏi:** Khối "repeat [5] times" làm gì?

**Các lựa chọn:**
- A) Lặp mãi mãi
- B) Lặp đúng 5 lần
- C) Lặp cho đến khi một điều kiện đúng
- D) Chỉ lặp nếu một biến bằng 5
- E) Không làm gì

**Đáp án: B**

---

### Giải thích chi tiết:

**1. Khối "repeat [N] times":**
- Chạy các khối bên trong nó **đúng N lần**, không hơn không kém
- N có thể là số cố định (5) hoặc biến

**2. So sánh với các vòng lặp khác:**
- "repeat **while** &lt;điều kiện&gt;" → lặp **khi** điều kiện còn đúng (có thể lặp 0, 1, hoặc nhiều lần)
- "repeat **until** &lt;điều kiện&gt;" → lặp **cho đến khi** điều kiện đúng
- "repeat [N] times" → lặp **đúng N lần**

**Ghi nhớ:** "repeat [5] times" = chạy 5 lần, không phụ thuộc điều kiện.

### Câu 3: Phép Cộng Cơ bản

**Câu hỏi:** Trong Blockly, kết quả của biểu thức [8] + [5] là gì?

**Các lựa chọn:** A) 85  B) 13  C) 3  D) 40  E) Error

**Đáp án: B**

---

### Giải thích chi tiết:

**1. Phép cộng số:**
- Khối [8] + [5] thực hiện **cộng hai số**: 8 + 5 = **13**
- Kết quả là một số (13), không phải chuỗi "85"

**2. Lưu ý:**
- Nếu dùng **join** với hai chuỗi "8" và "5" thì kết quả là "85" (nối chữ)
- Ở đây đề bài nói [8] + [5] là **phép cộng số** nên đáp án là 13

**Ghi nhớ:** [số] + [số] = cộng số; join [chữ] [chữ] = nối chuỗi.

### Câu 4: Khối If-Then-Else

**Câu hỏi:** Khối "if &lt;điều kiện&gt; then ... else ..." làm gì?

**Các lựa chọn:**
- A) Luôn chạy phần "then"
- B) Chạy phần "then" khi điều kiện **đúng**; chạy phần "else" khi điều kiện **sai**
- C) Chạy cả "then" và "else" mỗi lần
- D) Dừng chương trình
- E) Lặp điều kiện 10 lần

**Đáp án: B**

---

### Giải thích chi tiết:

**1. Cách hoạt động:**
- **Nếu điều kiện đúng (true):** chỉ chạy các khối trong nhánh **then**
- **Nếu điều kiện sai (false):** chỉ chạy các khối trong nhánh **else**
- Chỉ **một trong hai** nhánh được chạy mỗi lần

**2. Ví dụ:**
- Nếu `tuổi > 18` thì "Bạn đã trưởng thành" else "Bạn chưa trưởng thành"
- Khi tuổi = 20 → điều kiện đúng → in "Bạn đã trưởng thành"

**Ghi nhớ:** If-then-else = rẽ nhánh: đúng → then, sai → else; chỉ chạy một nhánh.

### Câu 5: Danh sách (List) là gì?

**Câu hỏi:** Trong Blockly, danh sách (list) là gì?

**Các lựa chọn:**
- A) Một số đơn
- B) Một tập hợp các phần tử (số, chữ, ...) theo **thứ tự**
- C) Một loại biến không thể thay đổi
- D) Một vòng lặp chạy ngược
- E) Một phép so sánh như &lt; hoặc &gt;

**Đáp án: B**

---

### Giải thích chi tiết:

**1. Khái niệm danh sách:**
- Danh sách chứa **nhiều giá trị** trong một biến, theo thứ tự (phần tử 1, 2, 3, ...)
- Có thể truy cập từng phần tử bằng **vị trí (chỉ số)**
- Blockly dùng **chỉ số bắt đầu từ 1**: phần tử đầu tiên là item 1

**2. Ví dụ:**
- Danh sách [10, 20, 30] → item 1 = 10, item 2 = 20, item 3 = 30

**Ghi nhớ:** List = nhiều giá trị có thứ tự; truy cập bằng item [vị trí] of [list].

### Câu 6: Khối Set (Gán giá trị)

**Câu hỏi:** "Set [x] to [10]" làm gì?

**Các lựa chọn:**
- A) Cộng 10 vào giá trị hiện tại của x
- B) Lưu giá trị 10 vào biến x (thay thế giá trị cũ)
- C) So sánh x với 10
- D) Xóa biến x
- E) In 10 ra màn hình

**Đáp án: B**

---

### Giải thích chi tiết:

**1. Set vs Change:**
- **Set [x] to [10]:** Đặt (gán) x **bằng** 10. Giá trị cũ của x bị thay thế.
- **Change [x] by [5]:** Cộng thêm 5 vào giá trị hiện tại của x (x = x + 5)

**2. Ví dụ:**
- Ban đầu x = 3. Sau "set [x] to [10]" → x = 10
- Ban đầu x = 3. Sau "change [x] by [5]" → x = 8

**Ghi nhớ:** Set = gán mới (thay thế); Change = cộng/trừ vào giá trị hiện có.

**Đoạn code:**
```
set [a] to [5]
change [a] by [3]
set [result] to [a]
```

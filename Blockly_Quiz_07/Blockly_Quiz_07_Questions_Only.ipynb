### Question 20: Nested Logic

```
set [score] to [85]
if &lt;[score] ≥ [90]&gt; then
    set [grade] to ["A"]
else
    if &lt;[score] ≥ [80]&gt; then
        set [grade] to ["B"]
    else
        set [grade] to ["C"]
set [result] to [grade]
```

What is the final value of `result`?

A) "A"  
B) "B"  
C) "C"  
D) 85  
E) 80
